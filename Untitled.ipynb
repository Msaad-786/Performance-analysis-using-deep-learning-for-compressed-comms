```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

# Parameters for Rayleigh distribution
scale_parameter = 1.0  # This can be adjusted as needed

# Generate random samples from Rayleigh distribution
num_samples = 100000
samples = rayleigh.rvs(scale=scale_parameter, size=num_samples)

# Plotting the PDF of Rayleigh distribution
x = np.linspace(0, 5, 1000)  # Range for plotting
pdf = rayleigh.pdf(x, scale=scale_parameter)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Rayleigh PDF', color='blue')
plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram of Samples')

plt.title('Rayleigh Fading PDF')
plt.xlabel('Signal Amplitude')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Save the plot as an image file
plt.savefig('rayleigh_fading_pdf.png', dpi=300)

# Display the plot
plt.show()
```
